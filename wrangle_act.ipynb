### Code

**1. Transpose (Make the prediction into one clumn instead of multiple columns) the Predicted fields to make the data cleaner and easier to use.**

**2. Remove wrong/False prediction (e.g. in one instance "paper Towel" was predicted as a dog breed.)**

**3. Remove Duplicates**

**4. Keep the highest Prediction confidence record for each Image. Drop rest of the records.**

```python
#Transpose P1, P2, P3 (Breed Predition) Columns into one column "pred_val"
img_prediction_clean_melt1 = pd.melt(img_prediction_clean, id_vars=['tweet_id', 'jpg_url', 'img_num'],
                           value_vars=['p1','p2','p3'], var_name='prediction', value_name='pred_val')

#Transpose 'p1_conf','p2_conf','p3_conf' (Predition confidence) Columns into one column "pred_conf"
img_prediction_clean_melt2 = pd.melt(img_prediction_clean, id_vars=['tweet_id', 'jpg_url', 'img_num'],
                           value_vars=['p1_conf','p2_conf','p3_conf'], var_name='prediction', value_name='pred_conf')
img_prediction_clean_melt2.prediction = img_prediction_clean_melt2.prediction.str[:2]

#Transpose 'p1_dog','p3_dog','p3_dog' (Predition truthfulness) Columns into one column "pred_accuracy"
img_prediction_clean_melt3 = pd.melt(img_prediction, id_vars=['tweet_id', 'jpg_url', 'img_num'],
                           value_vars=['p1_dog','p3_dog','p3_dog'], var_name='prediction', value_name='pred_accuracy')

img_prediction_clean_melt3.prediction = img_prediction_clean_melt3.prediction.str[:2]
```

```python
#Merge Three Melted Data Frames into one

img_prediction_clean_melt = img_prediction_clean_melt1.merge(img_prediction_clean_melt2,on=['tweet_id','img_num','jpg_url','prediction'])
img_prediction_clean_melt = img_prediction_clean_melt.merge(img_prediction_clean_melt3,on=['tweet_id','img_num','jpg_url','prediction'])
```

```python
#Test
img_prediction_clean_melt.head(5)
```

**2. Removing wrong/False prediction (e.g. in one instance "paper Towel" was predicted as a dog breed.)**

```python
img_prediction_clean_melt = img_prediction_clean_melt[img_prediction_clean_melt.pred_accuracy=='True']
img_prediction_clean_melt.shape
```
