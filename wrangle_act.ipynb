**1. Combine the 4 Dog stages fields into one and drop the 4 fields**

```python
# handle none
twitter_data_clean.doggo.replace('None', '', inplace=True)
twitter_data_clean.floofer.replace('None', '', inplace=True)
twitter_data_clean.pupper.replace('None', '', inplace=True)
twitter_data_clean.puppo.replace('None', '', inplace=True)

# merge into column
twitter_data_clean['dog_stage'] = twitter_data_clean.doggo + twitter_data_clean.floofer + twitter_data_clean.pupper + twitter_data_clean.puppo

# handle multiple stages
twitter_data_clean.loc[twitter_data_clean.dog_stage == 'doggopupper', 'dog_stage'] = 'doggo, pupper'
twitter_data_clean.loc[twitter_data_clean.dog_stage == 'doggopuppo', 'dog_stage'] = 'doggo, puppo'
twitter_data_clean.loc[twitter_data_clean.dog_stage == 'doggofloofer', 'dog_stage'] = 'doggo, floofer'

# handle missing values
twitter_data_clean.loc[twitter_data_clean.dog_stage == '', 'dog_stage'] = np.nan
```

**2. Drop Retweets and Replies from  Twitter_data and save it to "twitter_data_clean" dataframe**

```python
# Drop Retweets
twitter_data_clean = twitter_data_clean[~twitter_data.text.str.startswith("RT @")]
```

```python
# Drop Replies
twitter_data_clean = twitter_data_clean[~twitter_data_clean.in_reply_to_status_id.notnull()]
```

**Transpose the Predicted fields to make the data cleaner and easier to use**


```python
img_prediction_clean_melt1 = pd.melt(img_prediction, id_vars=['tweet_id', 'jpg_url', 'img_num'],
                           value_vars=['p1','p2','p3'], var_name='prediction', value_name='pred_val')

img_prediction_clean_melt2 = pd.melt(img_prediction, id_vars=['tweet_id', 'jpg_url', 'img_num'],
                           value_vars=['p1_conf','p2_conf','p3_conf'], var_name='prediction', value_name='pred_conf')

img_prediction_clean_melt2.prediction = img_prediction_clean_melt2.prediction.str[:2]
img_prediction_clean_melt3 = pd.melt(img_prediction, id_vars=['tweet_id', 'jpg_url', 'img_num'],
                           value_vars=['p1_dog','p3_dog','p3_dog'], var_name='prediction', value_name='pred_accuracy')

img_prediction_clean_melt3.prediction = img_prediction_clean_melt3.prediction.str[:2]
```

```python
img_prediction_clean_melt = img_prediction_clean_melt1.merge(img_prediction_clean_melt2,on=['tweet_id','img_num','jpg_url','prediction'])
img_prediction_clean_melt = img_prediction_clean_melt.merge(img_prediction_clean_melt3,on=['tweet_id','img_num','jpg_url','prediction'])
```

```python
#Remove duplicate records
img_prediction_clean_melt = img_prediction_clean_melt.drop_duplicates()
img_prediction_clean_melt.shape
```

```python
grouped = img_prediction_clean_melt.groupby(['tweet_id', 'jpg_url', 'img_num','prediction']).sum().reset_index()
grouped_sorted = grouped.sort_values('pred_conf', ascending=False)

img_prediction_clean_melted = grouped_sorted.groupby(['tweet_id', 'jpg_url', 'img_num']).first().reset_index()
img_prediction_clean_melted.shape
```
